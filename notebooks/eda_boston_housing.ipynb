```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações de visualização
sns.set(style="whitegrid")
plt.style.use("ggplot")

```

```python
# URL do conjunto de dados de Boston
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"

# Carregar os dados
data = pd.read_csv(url)

# Exibir as primeiras linhas do conjunto de dados
data.head()
```

```python
data.describe()
```

```python
# Distribuição de preços das casas
plt.figure(figsize=(10, 6))
sns.histplot(data['medv'], bins=30, kde=True)
plt.title('Distribuição de Preços das Casas')
plt.xlabel('MEDV')
plt.ylabel('Frequência')
plt.show()
```

```python
# Matriz de correlação
plt.figure(figsize=(12, 10))
correlation_matrix = data.corr().round(2)
sns.heatmap(data=correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlação')
plt.show()
```

```python
# Correlação com a variável alvo (medv)
correlation_target = correlation_matrix['medv'].sort_values(ascending=False)
correlation_target
```
